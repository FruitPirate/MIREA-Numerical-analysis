Решить задачу Коши методом Рунге-Кутты 4-го порядка со значениями шага 0.1, 0.2 и 0.01.  
y' = (-y*x + x*(x^2 + 1))/(x^2 + 1),  y(0) = 1,   0 <= x <= 1

```python
import numpy as np
import matplotlib.pyplot as plt
#x от 0 до 3

def f(x, y):
    return (-y*x + x*(x*x + 1))/(x*x + 1)

h1, h2, h3 = 0.1, 0.01, 0.2

x_arr1, x_arr2, x_arr3 = np.arange(0, 3+h1, h1), np.arange(0, 3+h2, h2), np.arange(0, 3+h3, h3)


def Runge_Kutta(arr_x, h):
    arr_y = np.zeros(len(arr_x))
    arr_y[0] = 1
    for i in range(1, len(arr_x)):
        k1 = f(arr_x[i-1], arr_y[i-1])
        k2 = f(arr_x[i-1] + h/2, arr_y[i-1] + k1 * h / 2)
        k3 = f(arr_x[i-1] + h/2, arr_y[i-1] + k2 * h / 2)
        k4 = f(arr_x[i-1] + h, arr_y[i-1] + k3 * h)
        arr_y[i] = arr_y[i-1] + h*(k1 + 2*k2 + 2*k3 + k4)/6
    return arr_y

y_arr1, y_arr2, y_arr3 = Runge_Kutta(x_arr1, h1), Runge_Kutta(x_arr2, h2), Runge_Kutta(x_arr3, h3)
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x_arr1, y_arr1, 'r')
plt.title('h = 0.1')
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x_arr2, y_arr2, 'g')
plt.title('h = 0.01')
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x_arr3, y_arr3, 'b')
plt.title('h = 0.2')
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x_arr1, y_arr1, 'r', x_arr2, y_arr2, 'g', x_arr3, y_arr3)
plt.legend(("h=0.1", "h=0.01", 'h=0.2'))
```
