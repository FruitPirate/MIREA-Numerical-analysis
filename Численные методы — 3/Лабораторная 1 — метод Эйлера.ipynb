Решить задачу Коши методом Эйлера и Эйлера с пересчёётом со значениями шага h1 = 0.1 и h2 = 0.01.  
y' = -y/(2*x) + x*x,   y(1) = 1,  1 <= x <= 2.

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x, y):
    return -y/(2*x) + x*x

h1, h2 = 0.1, 0.01
x_arr_h1 = np.arange(1, 2+h1, h1)
x_arr_h2 = np.arange(1, 2+h2, h2)
y_arr1_h1, y_arr2_h1 = np.zeros(x_arr_h1.size), np.zeros(x_arr_h1.size)
y_arr1_h2, y_arr2_h2 = np.zeros(x_arr_h2.size), np.zeros(x_arr_h2.size)
y_arr1_h1[0], y_arr1_h2[0], y_arr2_h1[0], y_arr2_h2[0] = 1, 1, 1, 1

#Шаг h = 0.1
for i in range(1, len(x_arr_h1)):
    #Метод Эйлера
    y_arr1_h1[i] = y_arr1_h1[i-1] + h1 * f(x_arr_h1[i-1], y_arr1_h1[i-1]) 
    #Метод Эйлера с пересчётом
    y_arr2_h1[i] = y_arr2_h1[i-1] + h1 * (f(x_arr_h1[i-1], y_arr2_h1[i-1]) + f(x_arr_h1[i], y_arr2_h1[i])) / 2 

#Шаг h = 0.01
for i in range(1, len(x_arr_h2)):
    y_arr1_h2[i] = y_arr1_h2[i-1] + h2 * f(x_arr_h2[i-1], y_arr1_h2[i-1])
    y_arr2_h2[i] = y_arr2_h2[i-1] + h2 * (f(x_arr_h2[i-1], y_arr2_h2[i-1]) + f(x_arr_h2[i], y_arr2_h2[i])) / 2

print("Для метода Эйлера с шагом h=0.1:", 'x = ', x_arr_h1, 'y = ', y_arr1_h1, sep='\n')
print("Для метода Эйлера с шагом h=0.01:", 'x = ', x_arr_h2, 'y = ', y_arr1_h2, sep='\n')
print("Для метода Эйлера с пересчётом с шагом h=0.1:", 'x = ', x_arr_h1, 'y = ', y_arr2_h1, sep='\n')
print("Для метода Эйлера с пересчётом с шагом h=0.01:", 'x = ', x_arr_h2, 'y = ', y_arr2_h2, sep='\n')


```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x_arr_h1, y_arr1_h1, 'r', x_arr_h2, y_arr1_h2, 'b')
plt.title("Сравнение методов Эйлера с разными шагами")
plt.legend(("Метод Эйлера с шагом 0.1", "Метод Эйлера с шагом 0.01"))
plt.show()
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x_arr_h1, y_arr2_h1, 'r', x_arr_h2, y_arr2_h2, 'b')
plt.title("Сравнение методов Эйлера с пересчётом с разными шагами")
plt.legend(("Метод Эйлера с пересчётом с шагом 0.1", "Метод Эйлера с пересчётом с шагом 0.01"))
plt.show()
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x_arr_h1, y_arr1_h1, 'r', x_arr_h1, y_arr2_h1, 'b')
plt.title("Сравнение метода Эйленра и метода Эйлера с пересчётом с шагом 0.1")
plt.legend(("Метод Эйлера с шагом 0.1", "Метод Эйлера с пересчётом с шагом 0.1"))
plt.show()
```

```python
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x_arr_h2, y_arr1_h2, 'r', x_arr_h2, y_arr2_h2, 'b')
plt.title("Сравнение метода Эйленра и метода Эйлера с пересчётом с шагом 0.01")
plt.legend(("Метод Эйлера с шагом 0.01", "Метод Эйлера с пересчётом с шагом 0.01"))
plt.show()
```
